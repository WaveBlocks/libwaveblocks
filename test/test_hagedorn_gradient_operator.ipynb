```python
import numpy
import math
import numpy.random as rand
import csv

from WaveBlocksND import *
from sample_wavepacket import *



print
print "--- Create sample wavepacket ---"

D = 4
shape = HyperbolicCutShape(D,7.0)
packet = createSampleWavepacket(D, shape)

print
print "PARAMETERS"
print packet.get_parameters()[0]
print packet.get_parameters()[1]
print packet.get_parameters()[2]
print packet.get_parameters()[3]

print
print "--- Compute wavepacket coefficients of gradient ---"
nabla = packet.get_gradient_operator()
shape_ext,gradient = nabla.apply_gradient_component(packet,0)

with open("gradient_coefficients.csv", "w") as filehandle:
    csvfile = csv.writer(filehandle, delimiter=' ')
    for node,entry in zip(list(shape_ext.get_node_iterator()),gradient):
        csvfile.writerow(node + tuple(numpy.real(entry)) + tuple(numpy.imag(entry)))

print
print "--- Evaluate gradient-wavepacket at a chosen location ---"
print
x0 = numpy.zeros((D,1))
for d in range(0,D):
    x0[d,0] = (d+1)/float(2*D)
print map(lambda wp: wp.slim_recursion(x0,0), nabla.apply_gradient(packet))

print
print "--- Evaluate gradient-wavepacket at some random locations ---"
gradpackets = nabla.apply_gradient(packet)

x0 = numpy.zeros((D,1))
for d in range(0,D):
    x0[d,0] = (d+1)/float(2*D)

rand.seed(0)
points = rand.randn(D,5)

results = numpy.array(map(lambda wp: wp.slim_recursion(points,0)[0], gradpackets))

with open("gradient_evaluation.csv", "w") as filehandle:
    csvfile = csv.writer(filehandle, delimiter=' ')
    for x,psi in zip(points.T,results.T):
        csvfile.writerow(tuple(x) + tuple(numpy.real(psi)) + tuple(numpy.imag(psi)))

print 
print "--- Done ---"
```
